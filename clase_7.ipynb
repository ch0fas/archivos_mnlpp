```python
import os
from dotenv import load_dotenv
from fredapi import Fred
import pandas as pd
import plotly.express as px
```

```python
load_dotenv()
```

```python
API_KEY = os.getenv("FRED_KEY")
#print(API_KEY)
```

```python
try:
    fred = Fred(API_KEY)

    series_id = "CSUSHPINSA"
    df = pd.DataFrame(fred.get_series(series_id), columns=['value'])
    df.index.name = 'date'

    #Datos desde el 2010 solamente
    df = df[df.index >= '2010-01-01']

    df.index = pd.DatetimeIndex(df.index).to_period('M').to_timestamp()
    df.index.freq = 'MS'

except Exception as e:
    print(e)
```

```python
df.head(15)
```

```python
fig = px.line(df, x=df.index, y='value', title=f'Serie de Tiempo: {series_id}')
fig.show()
```

```python
# Primer diferencial
df['diff_1'] = df['value'].diff()
```

```python
fig = px.line(df, x=df.index, y='diff_1', title=f'Serie de Tiempo Diferenciada: {series_id}')
fig.show()
```
